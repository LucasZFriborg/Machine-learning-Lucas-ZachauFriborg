## About
This notebook explores the MovieLens dataset, focusing on movies and tags to understand how they can be used in the movie recommendation system.

```python
import pandas as pd
```

## Movies exploration

```python
movies = pd.read_csv('../Data/ml-latest/movies.csv')
movies.head()
```

```python
movies.shape
```

```python
movies.info()
```

```python
movies['genres'].head()
```

Converting text to list:

```python
movies['genres'] = movies['genres'].apply(lambda x: x.split('|'))
```

Gathering all genres to see how many and to understand feature space (i.e., the space where each data point is represented by its features.
(dimensions = features, points = observations)):

```python
all_genres = set(g for genres in movies['genres'] for g in genres)
len(all_genres)
```

```python
all_genres
```

### Insights
- Movies may have multiple genres.
- The dataset is multi-label.
- Movies with missing genre ('no genres listed') may be filtered out from the dataset, since they do not provide useful information.
- Good for content-based filtering.
